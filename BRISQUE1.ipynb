```python
import cv2
import numpy as np
import imquality.brisque as brisque
from skimage import io, img_as_float
from skimage.util import random_noise
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('BUTTERFLY.png', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
img = img_as_float(io.imread('BUTTERFLY.png', as_gray=True))
```

```python

score = brisque.score(img)
print("Brisque score = ", score)
```

```python
 
# Load the image 
img = cv2.imread('BUTTERFLY.png')
 
# Add salt-and-pepper noise to the image.
noise_img = random_noise(img, mode='s&p',amount=0.3)
 
# The above function returns a floating-point image
# on the range [0, 1], thus we changed it to 'uint8'
# and from [0,255]
noise_img = np.array(255*noise_img, dtype = 'uint8')
 
cv2_imshow(noise_img)
```

```python

score = brisque.score(noise_img)
print("Brisque score = ", score)
```
