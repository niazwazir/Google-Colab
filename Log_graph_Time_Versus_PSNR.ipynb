```python
#AS A SAMPLE THIS GRAPH SAME AS 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import *
```

```python
psnr_scale_2 = [36.5,  36.5,  36.596,  36.68,  37.53]
no_of_parameters = [80,2,0.99,3,0.1]
Name_Of_Models = ["SelfEx","RFL","A+","SRCNN","VDSR (Ours)"]
for k in range(5):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_2[k]
    if loc_x == 0.1 and loc_y == 37.53:
        plt.semilogx(loc_x,loc_y,color='b',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=16,color='b')
    else:
        plt.semilogx(loc_x,loc_y,color='r',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12,color='r')
    
plt.ylim([36.4,37.6])
plt.xlim(100,0.0018)
#plt.xlim(10,10000)
plt.ylabel('PSNR(dB)',fontsize=16)
plt.xlabel('slow         running time(s)                fast',fontsize=12)
plt.grid(which="both")
plt.show()
```
