```python

# Creating dataset
x = ['SRCNN', 'FSRCNN','VDSR', 'DRCN', 'LapSRN', 'DRRN','SENext']
dataset_1 = [36.72,37.05,37.53,37.63,37.52,37.74,37.88]
dataset_2 = [0.955, 0.956, 0.959,0.959, 0.959, 0.959,0.9592]
 
# Creating plot with dataset_1
fig, ax1 = plt.subplots()
 
color = 'tab:red'
ax1.set_xlabel('State Of the Art Methods SET5 SCALE FACTOR 2X')
ax1.set_ylabel('PSNR (dB)', color = color)
ax1.plot(x, dataset_1,  'o-')
ax1.tick_params(axis ='y', labelcolor = color)
 
# Adding Twin Axes to plot using dataset_2
ax2 = ax1.twinx()
 
color = 'tab:blue'
ax2.set_ylabel('SSIM', color = color)
ax2.plot(x, dataset_2, 'xr-')
ax2.tick_params(axis ='y', labelcolor = color)
ax1.grid()
ax2.grid()
# Adding title
plt.title('ABALATION STUDY WITH SENext MODEL')
# Show plot
plt.show()
```

```python
psnr_scale_2 = [      36.54, 36.49,  36.72,  37.05, 36.58,  37.53, 37.52,  37.62,  37.80]
GPU_Processing_Time =[2,    80,      0.2,   0.009, 4,   0.9,   0.1,    0.9,    0.08]
Name_Of_Models = ["A+","SelfExSR","SRCNN","FSRCNN","SCN","VDSR","LapSRN","DCSCN","Ours"]
for k in range(9):
    loc_x = GPU_Processing_Time[k]
    loc_y = psnr_scale_2[k]
    if loc_x == 0.08 and loc_y == 37.80:
        plt.semilogx(loc_x,loc_y,color='b',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=14,color='b')
    else:
        plt.semilogx(loc_x,loc_y,color='r',marker='.')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=12,color='r')
    
plt.ylim([36.40,37.86])
plt.xlim(100,0.0020)
plt.ylabel('PSNR(dB)',fontsize=16)
plt.xlabel('slow         running time(s)                fast',fontsize=12)
plt.grid(which="both")
plt.show()
```
