```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
image = cv2.imread("bird_GT.bmp",1)
img_cvt=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.grid(None)
plt.show()
```

```
b,g,r = cv2.split (image)
```

```
plt.imshow(b)
plt.grid(None)
```

```
plt.imshow(g)
plt.grid(None)
```

```
plt.imshow(r)
plt.grid(None)
```

```
imgn=cv2.merge((b,g,r))
plt.imshow(imgn)
plt.grid(None)
```
