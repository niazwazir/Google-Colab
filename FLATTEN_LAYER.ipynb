**FLATTEN LAYER**
Flatten is the function that converts the pooled feature map to a single column or into single value that is passed to the fully connected layer.

```
from keras.layers import Input
from keras.layers import Conv2D
from keras.layers import Activation
from keras.layers import Flatten
```

```
image_size = (28,28,1)
input_img = Input(shape=(image_size))
C1 = C1 = Conv2D(64,(9,9),padding='SAME',name='CONV1')(input_img)
A1 = Activation('relu', name='act1')(C1)
C2 = Conv2D(32,(1,1),padding='SAME',name='CONV2')(A1)
A2 = Activation('relu', name='act2')(C2)
F = Flatten()(A2)
```

```
from keras.models import Model
model = Model(input_img, F)
model.summary()
```
