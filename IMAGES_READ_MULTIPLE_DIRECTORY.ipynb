```
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from glob import glob
```

```
cleanPath = glob('Clean_Images/*.*')
blurryPath = glob('Blurry_Images/*.*')
Images = [];Blurry = []
for image, blurry in zip (cleanPath, blurryPath):
    Images.append(imread(image))
    Blurry.append(imread(blurry))
Images = np.array(Images).astype('float32')
Blurry = np.array(Blurry).astype('float32')
```

```
#SIZE OF IMAGES IN BOTH DIRECTORY ARE SAME OTHERWISE CODE CAN NOT RUN
f,ax = plt.subplots(2,5,figsize=(25,5))
for i in range(5):
    ax[0,i].imshow(Images[i].astype("uint8")); ax[0,i].axis('Off');ax[0,i].set_title('ModelImage',size=20)
    ax[1,i].imshow(Blurry[i].astype("uint8")); ax[1,i].axis('Off');ax[1,i].set_title('BicubicImage',size=20)
plt.show()
```
