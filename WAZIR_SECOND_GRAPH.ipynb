```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

psnr_scale_2      = [36.93,   36.66,   37.62,   37.74,  37.17,  37.53,  37.78,    37.63,  37.74,      37.66,    37.80   ]
no_of_parameters  = [42,      57,      244,     297,    594,    665,    677,      1774,   2550,       4131,     245     ]
Name_Of_Models    = ["SCN",  "SRCNN", "DCSCN", "DRRN", "SPSR", "VDSR", "Memnet", "DRCN", "DenseNet", "RED30", "Proposed"]

plt.grid(True)

for k in range(11):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_2[k]
    plt.semilogx(loc_x,loc_y,color='b', marker='D')
    plt.text(loc_x+3,loc_y,str(Name_Of_Models[k]),fontsize=12)

plt.ylim([36.60,37.9])
plt.xlim([10,10000])
plt.ylabel('PSNR(dB)',fontsize=12)
plt.xlabel('Smaller <-------------------------Number of Parameters (K)------------------------->Larger',fontsize=12)
#plt.xlabel('Slower<-   Running time(sec)  ->Faster',fontsize=12)

plt.show()


```
