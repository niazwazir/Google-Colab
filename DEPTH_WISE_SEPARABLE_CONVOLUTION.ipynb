```
from keras.layers import Input, Conv2D, Activation, SeparableConv2D
from keras.models import Model
from keras import optimizers
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
```

```
img_shape = (32,32,1)
input_img = Input(shape=(img_shape))
C1 = Conv2D(64,(9,9),padding='SAME',name='CONV1')(input_img)
A1 = Activation('relu', name='act1')(C1)
C2 = Conv2D(32,(1,1),padding='SAME',name='CONV2')(A1)
A2 = Activation('relu', name='act2')(C2)
C3 = Conv2D(1,(5,5),padding='SAME',name='CONV3')(A2)
A3 = Activation('relu', name='act3')(C3)
model = Model(input_img, A3)
opt = optimizers.Adam(lr=0.0003)
model.compile(optimizer=opt,loss='mean_squared_error')
model.summary()
```

```
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```

```
img_shape = (32,32,1)
input_img = Input(shape=(img_shape))
C1 = SeparableConv2D(64,(9,9),padding='SAME')(input_img)
A1 = Activation('relu', name='act1')(C1)
C2 = SeparableConv2D(32,(1,1),padding='SAME')(A1)
A2 = Activation('relu', name='act2')(C2)
C3 = SeparableConv2D(1,(5,5),padding='SAME')(A2)
A3 = Activation('relu', name='act3')(C3)
model1 = Model(input_img, A3)
opt = optimizers.Adam(lr=0.0003)
model1.compile(optimizer=opt,loss='mean_squared_error')
model1.summary()
```

```
SVG(model_to_dot(model1, show_shapes=True).create(prog='dot', format='svg'))
```
