```
import numpy as np
import cv2
import matplotlib.pyplot as plt
img1 = cv2.imread('zebra.png')
img=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
cv2.rectangle(img,(20,150),(100,250),(255,0,0),2) # X-AXIS in B/W 20 and 100, Y-Axis in B/W 150 and 250
plt.imshow(img)
plt.grid(0)
plt.show()
```
