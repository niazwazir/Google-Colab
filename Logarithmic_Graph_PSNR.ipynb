```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib.pyplot import *
```

```python
psnr_scale_2 = [36.66,  37.53,  37.66,  37.38,  37.80,  37.62]
no_of_parameters = [57,665,4131,677,240,244]
Name_Of_Models = ["SRCNN","VDSR","RED","Memnet","Proposed","DCSCN"]
for k in range(6):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_2[k]
    plt.semilogx(loc_x,loc_y,color='b',marker='.')
    plt.text(loc_x+3,loc_y,str(Name_Of_Models[k]),fontsize=12,color='r')
    
plt.ylim([36,38])
plt.xlim(10,10000)
plt.ylabel('PSNR(dB)',fontsize=12)
plt.xlabel('Smaller <------------Number of parameters (K)------------>Larger',fontsize=12)
plt.grid(which="both")
plt.show()
```
