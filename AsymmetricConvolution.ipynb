```
from keras.models import Sequential
from keras.layers import Conv2D
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot
```

**3 x 3 Convolution Layer**

```
model3 = Sequential([
    Conv2D(10, kernel_size=(3,3), input_shape=(28,28,10), activation ='relu',padding='same', use_bias=False),
    
])
model3.summary()
```

```
SVG(model_to_dot(model3, show_shapes=True).create(prog='dot', format='svg'))
```

**3 x 1 and 1 x 3 Convolution Layers**

```
model31 = Sequential([
    Conv2D(10, kernel_size=(3,1), input_shape=(28,28,10), activation ='relu',padding='same', use_bias=False),
    Conv2D(10, kernel_size=(1,3), activation ='relu',padding='same', use_bias=False),
    
])
model31.summary()
```

```
SVG(model_to_dot(model31, show_shapes=True).create(prog='dot', format='svg'))
```

**5 x 5 Convolution Layer**

```
SVG(model_to_dot(model5, show_shapes=True).create(prog='dot', format='svg'))
```

**5 x 1 and 1 x 5 Convolution Layer**

```
model51 = Sequential([
    Conv2D(10, kernel_size=(5,1), input_shape=(28,28,10), activation ='relu',padding='same', use_bias=False),
    Conv2D(10, kernel_size=(1,5), activation ='relu',padding='same', use_bias=False),
    
])
model51.summary()
```

```
SVG(model_to_dot(model51, show_shapes=True).create(prog='dot', format='svg'))
```
