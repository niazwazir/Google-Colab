```
#FSRCNN MODEL
# You can freely modify this file.
# However, you need to have a function that is named get_model and returns a Keras Model.

import keras as k
from keras import models
from keras import layers
from keras import utils
from keras.layers import Conv2DTranspose
from keras.layers import Dense, Conv2D, Activation

def get_model():
    img_height = 32
    img_width = 32
    img_channels = 1

    input_shape = (img_height, img_width, img_channels)
    img_input = k.Input(shape=input_shape)
    C1 = layers.Conv2D(56, 5, padding = 'same', kernel_initializer = 'he_normal')(img_input)
    A1 = Activation('relu')(C1)
    C2 = layers.Conv2D(12, 1, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A1)
    A2 = Activation('relu')(C2)
    C3 = layers.Conv2D(12, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A2)
    A3 = Activation('relu')(C3)
    
    C4 = layers.Conv2D(12, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A3)
    A4 = Activation('relu')(C4)
    
    C5 = layers.Conv2D(12, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A4)
    A5 = Activation('relu')(C5)
    
    C6 = layers.Conv2D(12, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A5)
    A6 = Activation('relu')(C6)
    D1 = Conv2DTranspose(1,9, strides=(2,2),padding='same')(A6)
    model = models.Model(img_input, D1)

    return model
```

```
#VDSR
# You can freely modify this file.
# However, you need to have a function that is named get_model and returns a Keras Model.

import keras as k
from keras import models
from keras import layers
from keras import utils
from keras.layers import Conv2DTranspose
from keras.layers.merge import concatenate
from keras.layers import Dense, Conv2D, Activation

def get_model():
    img_height = 41
    img_width = 41
    img_channels = 1

    input_shape = (img_height, img_width, img_channels)
    img_input = k.Input(shape=input_shape)
    C1 = layers.Conv2D(64, 3, padding = 'same', kernel_initializer = 'he_normal')(img_input)
    A1 = Activation('relu')(C1)
    C2 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A1)
    A2 = Activation('relu')(C2)
    C3 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A2)
    A3 = Activation('relu')(C3)
    
    C4 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A3)
    A4 = Activation('relu')(C4)
    
    C5 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A4)
    A5 = Activation('relu')(C5)
    
    C6 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A5)
    A6 = Activation('relu')(C6)
    
    C7 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A6)
    A7 = Activation('relu')(C7)
    
    C8 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A7)
    A8 = Activation('relu')(C8)
    
    C20 = layers.Conv2D(1, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(A8)
    A20 = Activation('relu')(C20)
    
    A22 = concatenate(inputs=[img_input,A20])
    
    
    model = models.Model(img_input, A22)

    return model
```

```
import keras as k
from keras import models
from keras import layers
from keras import utils
from keras.layers import Dense, Conv2D, Activation

def get_model():
    img_height = 32
    img_width = 32
    img_channels = 3

    input_shape = (img_height, img_width, img_channels)
    img_input = k.Input(shape=input_shape)
    #FIRST FEATURE EXTRACTION LAYER
    C1 = layers.Conv2D(64, 3, padding = 'same', kernel_initializer = 'he_normal')(img_input)
    A1 = layers.LeakyReLU(alpha=0.3)(C1)
    
    #FIRST RESNET BLOCK
    C2 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(A1)
    A2 = layers.LeakyReLU(alpha=0.3)(C2)
    
    C3 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(A2)
    A3 = layers.LeakyReLU(alpha=0.3)(C3)
    Sum1  = layers.add([A1,A3])
    
    #SECOND RESNET BLOCK
    C4 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(Sum1)
    A4 = layers.LeakyReLU(alpha=0.3)(C4)
    
    C5 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(A4)
    A5 = layers.LeakyReLU(alpha=0.3)(C5)
    Sum2  = layers.add([A1,A5])
    
    #SHRINKING LAYER
    C6 = layers.SeparableConv2D(64, 1,padding = 'same', kernel_initializer = 'he_normal')(Sum2)
    A6 = layers.LeakyReLU(alpha=0.3)(C6)
    
    #DECONVOLUTION LAYER
    y = layers.Conv2DTranspose(3,9, strides=(2,2),padding='same')(A6)
    
    #EXPANDING LAYER 
    C7 = layers.SeparableConv2D(64, 1,padding = 'same', kernel_initializer = 'he_normal')(y)
    A7 = layers.LeakyReLU(alpha=0.3)(C7)
    #XCEPTION BLOCK 1
    X1 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(A7)
    AX1 = layers.LeakyReLU(alpha=0.3)(X1)
    
    X2 = layers.SeparableConv2D(64, 5,padding = 'same', kernel_initializer = 'he_normal')(A7)
    AX2 = layers.LeakyReLU(alpha=0.3)(X2)
    
    X3 = layers.SeparableConv2D(64, 7,padding = 'same', kernel_initializer = 'he_normal')(A7)
    AX3 = layers.LeakyReLU(alpha=0.3)(X3)
    
    Sum3  = layers.add([AX1,AX2,AX3])
    
    C8 = layers.SeparableConv2D(3,3,padding = 'same', kernel_initializer = 'he_normal')(Sum3)
    A8 = layers.LeakyReLU(alpha=0.3)(C8)
    
    #XCEPTION BLOCK 2
    X4 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(A8)
    AX4 = layers.LeakyReLU(alpha=0.3)(X4)
    
    X5 = layers.SeparableConv2D(64, 5,padding = 'same', kernel_initializer = 'he_normal')(A8)
    AX5 = layers.LeakyReLU(alpha=0.3)(X5)
    
    X6 = layers.SeparableConv2D(64, 7,padding = 'same', kernel_initializer = 'he_normal')(A8)
    AX6 = layers.LeakyReLU(alpha=0.3)(X6)
    Sum4  = layers.add([AX4,AX5,AX6])
    C9 = layers.SeparableConv2D(3,3,padding = 'same', kernel_initializer = 'he_normal')(Sum4)
    A9 = layers.LeakyReLU(alpha=0.3)(C9)
    #XCEPTION BLOCK 3
    X7 = layers.SeparableConv2D(64, 3,padding = 'same', kernel_initializer = 'he_normal')(A9)
    AX7 = layers.LeakyReLU(alpha=0.3)(X7)
    
    X8 = layers.SeparableConv2D(64, 5,padding = 'same', kernel_initializer = 'he_normal')(A9)
    AX8 = layers.LeakyReLU(alpha=0.3)(X8)
    
    X9 = layers.SeparableConv2D(64, 7,padding = 'same', kernel_initializer = 'he_normal')(A9)
    AX9 = layers.LeakyReLU(alpha=0.3)(X9)
    Sum5  = layers.add([AX7,AX8,AX9])
    
    C10 = layers.SeparableConv2D(3,3,padding = 'same', kernel_initializer = 'he_normal')(Sum5)
    A10 = layers.LeakyReLU(alpha=0.3)(C10)


    model = models.Model(img_input, A10)

    return model
```

```
# You can freely modify this file.
# However, you need to have a function that is named get_model and returns a Keras Model.
import keras as k
from keras import models
from keras import layers
from keras import utils

def get_model():
    img_height = 256
    img_width = 256
    img_channels = 1

    input_shape = (img_height, img_width, img_channels)
    img_input = k.Input(shape=input_shape)
    conv1 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(img_input)
    conv1 = layers.Conv2D(64, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv1)
    pool1 = layers.MaxPooling2D(pool_size=(2, 2))(conv1)
    conv2 = layers.Conv2D(128, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(pool1)
    conv2 = layers.Conv2D(128, 3, activation = 'relu', padding = 'same', kernel_initializer = 'he_normal')(conv2)
    pool2 = layers.MaxPooling2D(pool_size=(2, 2))(conv2)

    model = models.Model(img_input, pool2)

    return model
```
