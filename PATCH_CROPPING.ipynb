```
from PIL import Image 
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
img1 = cv2.imread("Cropped_Image.png")
img_cvt=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.show()
```
