```
from keras.layers import Input
from keras.layers import Conv2DTranspose
from keras.layers import Conv2D
from keras.models import Model
```

```
img_shape = (256,256,3)
input_img = Input(shape=(img_shape))
C1 = Conv2D(6,(2,2),strides=(2,2),padding='SAME',name='CONV1')(input_img)
D1 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV1')(C1)
C2 = Conv2D(6,(2,2),strides=(2,2),padding='SAME',name='CONV2')(D1)
D2 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV2')(C2)
C3 = Conv2D(6,(2,2),strides=(2,2),padding='SAME',name='CONV3')(D2)
D3 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV3')(C3)
C4 = Conv2D(6,(2,2),strides=(2,2),padding='SAME',name='CONV4')(D3)
D4 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV4')(C4)
model = Model(input_img, D4)
model.summary()
```
