```
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```
image1 = cv2.imread("baby.bmp")
img_cvt=cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
plt.imshow(img_cvt)
plt.grid(None)
plt.show()
```

```
b = image1.copy()
# set green and red channels to 0  # b = 0, g = 1 and r = 2
b[:, :, 1] = 0
b[:, :, 2] = 0
```

```
plt.imshow(b)
print(b.shape)
plt.grid(None)
plt.show()
```

```
g = image1.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0
```

```
plt.imshow(g)
print(g.shape)
plt.grid(None)
plt.show()
```

```
r = image1.copy()
# set blue and green channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0
```

```
plt.imshow(r)
print(r.shape)
plt.grid(None)
plt.show()
```

```
add = b + g + r 
plt.imshow(add)
plt.grid(None)
```

```
RGB_img = cv2.cvtColor(add, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.grid(None)
plt.show()
```
