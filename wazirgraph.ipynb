```
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

psnr_scale_2      = [36.66,   37.00,     37.53,     37.66,          37.78,     37.51,  37.62]
no_of_parameters  = [57,      12,        665,       4131,             677,       240,  244]
Name_Of_Models              = ["SRCNN", "FSRCNN",  "VDSR",    "RED",      "Memnet",    "Proposed", "DCSCN"]

plt.grid(True)

for k in range(7):
    loc_x = no_of_parameters[k]
    loc_y = psnr_scale_2[k]
    plt.semilogx(loc_x,loc_y,color='b', marker='D')
    plt.text(loc_x+3,loc_y,str(Name_Of_Models[k]),fontsize=12)

plt.ylim([36,38])
plt.xlim([1,10000])
plt.ylabel('PSNR(dB)',fontsize=12)
plt.xlabel('Smaller <-------------------------Number of Parameters (K)------------------------->Larger',fontsize=12)
#plt.xlabel('Slower<-   Running time(sec)  ->Faster',fontsize=12)

plt.show()
```
