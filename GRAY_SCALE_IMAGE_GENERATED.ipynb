```
import numpy as np
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('zebra.png')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize = (14,14), dpi = 100)
ax = plt.subplot("121")
ax.imshow(img)
ax.set_title("GT")
plt.grid(0)
plt.show()
```

```
img = cv2.imread('zebra.png',cv2.IMREAD_GRAYSCALE)
fig = plt.figure(figsize = (14,14), dpi = 100)
ax = plt.subplot("121")
ax.imshow(img)
ax.set_title("GT")
plt.grid(0)
plt.show()
```
