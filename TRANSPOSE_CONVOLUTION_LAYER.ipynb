```
from keras.layers import Input
from keras.layers import Conv2DTranspose
from keras.layers import Conv2D
from keras.models import Model
```

```
img_shape = (32,32,1)
input_img = Input(shape=(img_shape))
D1 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV1')(input_img)
C1 = Conv2D(32,(1,1),padding='SAME',name='CONV1')(D1)
D2 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV2')(C1)
C2 = Conv2D(32,(1,1),padding='SAME',name='CONV2')(D2)
D3 = Conv2DTranspose(32, (3, 3), strides=(2,2), padding='same',name='TRANS_CONV3')(C2)
model = Model(input_img, D3)
model.summary()
```
